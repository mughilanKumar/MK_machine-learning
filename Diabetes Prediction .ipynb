```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
print(os.listdir())
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.info()
```

```python
df = df.replace(0 ,np.NaN)
df.loc[:,('Pregnancies','Outcome')] = df.loc[:,('Pregnancies','Outcome')].replace(np.NaN,0)
df.info()
```

```python
df['Glucose']= df['Glucose'].replace(np.NaN,df['Glucose'].mean())
df['BMI']= df['BMI'].replace(np.NaN,df['BMI'].mean())
df['BloodPressure']= df['BloodPressure'].replace(np.NaN,df['BloodPressure'].mean())
df['SkinThickness']= df['SkinThickness'].replace(np.NaN,df['SkinThickness'].mean())
df.info()
df.corr()
```

```python
df_wn = df.dropna()
df_wn.info()
df_wn.corr()
```

```python
ins_df = df_wn.loc[:,('Glucose','Insulin')]
ins_df.info()
```

```python
ins = ins_df.iloc[:,-1].values
gl = ins_df.iloc[:,0].values
np.corrcoef(ins,gl)
```

```python
plt.scatter(gl,ins)
```

```python
ins = ins.reshape(-1,1)
gl = gl.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
regressor = LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(gl,ins,test_size=0.25)
regressor.fit(x_train,y_train)
predicted = regressor.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,predicted)
```

```python
regressor.fit(ins,gl)
test_df = df.iloc[:,1].values
test_df= test_df.reshape(-1,1)
insuline_predict = regressor.predict(test_df)
#insuline_predict = insuline_predict.reshape(1,-1)
```

```python
insuline_predict
```

```python
insuline_predict = pd.DataFrame(insuline_predict)
```

```python
df.head()
```

```python
df['Insulin']= df['Insulin'].replace(to_replace=np.NaN, value= insuline_predict.loc[:,0])
```

```python
df.info()
```

```python
df.corr()
```

```python
x = df.iloc[:,1].values
x = x.reshape(-1,1)
y = df.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred,y_test)
accuracy = (cm[0,0]+cm[1,1])/(cm[0,0]+cm[0,1]+cm[1,0]+cm[1,1])
accuracy
```
